Matplotlib é uma biblioteca de software para criação de gráficos e visualizações de dados em geral, feita para e da linguagem de programação Python e sua extensão de matemática NumPy.

```python
# Dados para o gráfico de linha
x = [1, 2, 3, 4, 5]
```

```python
y = [2, 4, 6, 8, 10]
```

```python
# Criar o gráfico de linha
plt.plot(x, y)
```

```python
# Adicionar rótulos aos eixos x e y
plt.xlabel('Eixo X')
```

```python
plt.ylabel('Eixo Y')
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Dados para o gráfico de linha
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# Criar o gráfico de linha
plt.plot(x, y)
# Adicionar rótulos aos eixos x e y
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
# Adicionar título ao gráfico
plt.title('Gráfico de Linha Simples')
```

```python
import pandas as pd

df = pd.DataFrame({'mass': [0.330, 4.87 , 5.97],'radius': [2439.7, 6051.8, 6378.1]}, index=['Mercury', 'Venus', 'Earth'])

print(df)
plot = df.plot.pie(y='mass', figsize=(5, 5))
```
